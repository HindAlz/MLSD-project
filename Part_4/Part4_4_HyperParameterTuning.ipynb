```python
best_score = 0
best_params = None

for num_leaves in [31, 63]:
    for lr in [0.05, 0.1]:
        for n_est in [200, 500]:
            model = LGBMClassifier(
            num_leaves=num_leaves,
            learning_rate=lr,
            n_estimators=n_est,
            scale_pos_weight=4,  
            random_state=10
            )

            results = m.evaluate_model(X, y, f"LGBM | leaves={num_leaves} lr={lr}", model, preprocess_fn=m.apply_random_oversample)

            if results["recall"] > best_score:
                best_score = results["recall"]
                best_params = model.get_params()

print("Best F1:", best_score)
print("Best params:", best_params)
```

```python
from sklearn.ensemble import RandomForestClassifier

best_score = 0
best_params = None

for n_estimators in [200, 300, 500]:
    for max_depth in [None, 10, 15, 20]:
        for min_samples_leaf in [1, 2, 4]:

            model = RandomForestClassifier(
                n_estimators=n_estimators,
                max_depth=max_depth,
                min_samples_leaf=min_samples_leaf,
                random_state=10,
                n_jobs=-1
            )

            params = {
                "n_estimators": n_estimators,
                "max_depth": max_depth,
                "min_samples_leaf": min_samples_leaf
            }

            results = m.evaluate_model(X, y, f"RF | {params}", model, preprocess_fn=m.apply_random_oversample)

            if results["recall"] > best_score:
                best_score = results["recall"]
                best_params = params

print("Best recall:", best_score)
print("Best params:", best_params)
```

```python
from sklearn.neural_network import MLPClassifier

best_score = 0
best_params = None

for layers in [(64,), (128,), (64,32)]:
    for alpha in [1e-4, 1e-3]:
        for lr in [0.001, 0.01]:

            model = MLPClassifier(
                hidden_layer_sizes=layers,
                alpha=alpha,
                learning_rate_init=lr,
                max_iter=700,
                random_state=10
            )

            results = m.evaluate_model(X, y, f"MLP | layers={layers}, alpha={alpha}, lr={lr}", model, preprocess_fn=m.apply_smote)

            m.log_model_run(
                f"MLP | layers={layers}, alpha={alpha}, lr={lr}",
                model, results, X, y)

            if results["recall"] > best_score:
                best_score = results["recall"]
                best_params = {
                    "layers": layers,
                    "alpha": alpha,
                    "lr": lr
                }

print("Best recall:", best_score)
print("Best params:", best_params)
```
