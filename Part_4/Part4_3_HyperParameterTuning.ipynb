```python
from itertools import product

best_score = 0
best_params = None

for num_leaves, lr, n_est in product(
    [31, 63],
    [0.05, 0.1],
    [200, 500]
):
    model = LGBMClassifier(
        num_leaves=num_leaves,
        learning_rate=lr,
        n_estimators=n_est,
        scale_pos_weight=4,  
        random_state=10
    )

    results = m.evaluate_model(
        X,
        y,
        f"LGBM | leaves={num_leaves} lr={lr}",
        model,
        preprocess_fn=m.apply_random_oversample
    )

    if results["recall"] > best_score:
        best_score = results["recall"]
        best_params = model.get_params()

print("Best F1:", best_score)
print("Best params:", best_params)
```

```python
from itertools import product
from sklearn.ensemble import RandomForestClassifier

best_score = 0
best_params = None

param_grid = [
    {"n_estimators": 200, "max_depth": None, "min_samples_leaf": 1},
    {"n_estimators": 300, "max_depth": 10,   "min_samples_leaf": 1},
    {"n_estimators": 300, "max_depth": 15,   "min_samples_leaf": 2},
    {"n_estimators": 500, "max_depth": None, "min_samples_leaf": 2},
    {"n_estimators": 500, "max_depth": 20,   "min_samples_leaf": 4},
]

for params in param_grid:
    model = RandomForestClassifier(
        n_estimators=params["n_estimators"],
        max_depth=params["max_depth"],
        min_samples_leaf=params["min_samples_leaf"],
        random_state=10,
        n_jobs=-1
    )

    results = m.evaluate_model(
        X,
        y,
        f"RF | {params}",
        model,
        preprocess_fn=m.apply_random_oversample
    )

    if results["recall"] > best_score:
        best_score = results["recall"]
        best_params = params

print("Best recall:", best_score)
print("Best params:", best_params)
```

```python
from itertools import product
from sklearn.neural_network import MLPClassifier

best_score = 0
best_params = None

for layers, alpha, lr in product(
    [(64,), (128,), (64,32)],
    [1e-4, 1e-3],
    [0.001, 0.01]
):
    model = MLPClassifier(
        hidden_layer_sizes=layers,
        alpha=alpha,
        learning_rate_init=lr,
        max_iter=700,
        random_state=10
    )

    results = m.evaluate_model(
        X,
        y,
        f"MLP | layers={layers}, alpha={alpha}, lr={lr}",
        model,
        preprocess_fn=m.apply_smote   
    )
    m.log_model_run(f"MLP | layers={layers}, alpha={alpha}, lr={lr}", model, results, X, y)

    if results["recall"] > best_score:
        best_score = results["recall"]
        best_params = {
            "layers": layers,
            "alpha": alpha,
            "lr": lr
        }

print("Best recall:", best_score)
print("Best params:", best_params)
```

```python
import pandas as pd
from autogluon.tabular import TabularPredictor

# -----------------------------
# Paths
# -----------------------------
TRAIN_PATH = r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\train.parquet"
TEST_PATH = r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\val.parquet"
TARGET = "fake"

# Where AutoGluon will save everything
AUTOML_PATH = r"C:\Users\Hzaab\Desktop\MLSD project\scratch\autogluon_recall"

# -----------------------------
# Load data
# -----------------------------
train_df = pd.read_parquet(TRAIN_PATH)
test_df = pd.read_parquet(TEST_PATH)

# -----------------------------
# Train AutoML
# -----------------------------
predictor = TabularPredictor(
    label=TARGET,
    eval_metric="recall",
    path=AUTOML_PATH,
    verbosity=2,
).fit(
    train_data=train_df,
    presets="best_quality",
    time_limit=3600,   # 1 hour; change as needed
)

# -----------------------------
# Compare models
# -----------------------------
leaderboard = predictor.leaderboard(test_df, silent=True)
print("\n=== Leaderboard ===")
print(leaderboard)

# -----------------------------
# Evaluate on held-out test set
# -----------------------------
test_scores = predictor.evaluate(test_df)
print("\n=== Test metrics ===")
print(test_scores)

# -----------------------------
# Best model predictions
# -----------------------------
y_pred = predictor.predict(test_df.drop(columns=[TARGET]))
print("\nSample predictions:")
print(y_pred.head())

# Optional probabilities
y_prob = predictor.predict_proba(test_df.drop(columns=[TARGET]))
print("\nSample predicted probabilities:")
print(y_prob.head())
```
