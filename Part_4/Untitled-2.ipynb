```python
from sklearn.ensemble import VotingClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=10)
lgbm = LGBMClassifier(random_state=10)

ensemble = VotingClassifier(
    estimators=[
        ("rf", rf),
        ("lgbm", lgbm)
    ],
    voting="soft"   # IMPORTANT (uses probabilities)
)
```

```python
scores = m.evaluate_model(
    X,
    y,
    "RF + LGBM (SMOTE)",
    ensemble,
    preprocess_fn=m.apply_smote
)
```

```python
m.log_model_run("LGBM+RF_voting", ensemble, scores, X, y)
```

```python
from autogluon.tabular import TabularPredictor

train_df=pd.read_parquet(r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\train.parquet")
X_train = train_df.drop(columns=["fake"])
y_train = train_df["fake"]

predictor = TabularPredictor.load(
    r"C:\Users\Hzaab\Desktop\MLSD project\scratch\autogluon_auc"   # or gpu path
)
y_prob = predictor.predict_proba(X_train, model="LightGBM_r188_BAG_L2")[1]
```

```python
import numpy as np
from sklearn.metrics import f1_score

best_f1 = 0
best_t = 0.5

for t in np.linspace(0.1, 0.9, 50):
    y_pred = (y_prob >= t).astype(int)
    f1 = f1_score(y_train, y_pred)

    if f1 > best_f1:
        best_f1 = f1
        best_t = t

print("Best threshold:", best_t)
```

```python

test_df=pd.read_parquet(r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\val.parquet")
X_test = test_df.drop(columns=["fake"])
y_test = test_df["fake"]

y_prob_test = predictor.predict_proba(X_test)[1]
y_pred_test = (y_prob_test >= best_t).astype(int)
```

```python
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score

print("Accuracy:", accuracy_score(y_test, y_pred_test))
print("F1:", f1_score(y_test, y_pred_test))
print("Recall:", recall_score(y_test, y_pred_test))
print("Precision:", precision_score(y_test, y_pred_test))
```
