```python
from sklearn.ensemble import VotingClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=10)
lgbm = LGBMClassifier(random_state=10)

ensemble = VotingClassifier(
    estimators=[
        ("rf", rf),
        ("lgbm", lgbm)
    ],
    voting="soft"   # IMPORTANT (uses probabilities)
)
```

```python
train_df = pd.read_parquet('C:\\Users\\Hzaab\\Desktop\\MLSD project\\data\\preprocessed\\train.parquet', engine='fastparquet')
TARGET_COL = 'fake'

X = train_df.drop(columns=[TARGET_COL])
y = train_df[TARGET_COL]

cv = KFold(n_splits=5, shuffle=True, random_state=10)

from sklearn.model_selection import cross_validate
```

```python
scores = m.evaluate_model(
    X,
    y,
    "RF + LGBM (SMOTE)",
    ensemble,
    preprocess_fn=m.apply_smote
)
```

```python
m.log_model_run("LGBM+RF_voting", ensemble, scores, X, y)
```

```python
ensemble = VotingClassifier(
    estimators=[
        ("rf", rf),
        ("lgbm", lgbm)
    ],
    voting="soft",
    weights=[2, 1]   # LGBM stronger
)
```

```python
scores = m.evaluate_model(
    X,
    y,
    "RF + LGBM (SMOTE)",
    ensemble,
    preprocess_fn=m.apply_smote
)
```
