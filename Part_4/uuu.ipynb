```python
import pandas as pd
from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split

TARGET = "fake"
train_df=pd.read_parquet(r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\train.parquet")

train_sub, val_sub = train_test_split(
    train_df,
    test_size=0.2,
    stratify=train_df["fake"],
    random_state=10
)

train_sub=pd.DataFrame(train_sub)
val_sub=pd.DataFrame(val_sub)
train_sub.to_parquet(
    r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\train_sub.parquet",
    index=False
)

val_sub.to_parquet(
    r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\val_sub.parquet",
    index=False
)

```

```python


AUTOML_PATH = r"C:\Users\Hzaab\Desktop\MLSD project\huhi"

predictor = TabularPredictor(
    label="fake",
    eval_metric="roc_auc",   # 👈 keep this
    path=AUTOML_PATH,
    verbosity=2,
).fit(
    train_data=train_sub,
    tuning_data=val_sub,
    presets="best_quality",
    time_limit=3600,
)
```

```python


# -----------------------------
# Paths
# -----------------------------
TRAIN_PATH = r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\train_sub.parquet"
TEST_PATH = r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\val_sub.parquet"
TARGET = "fake"

# Where AutoGluon will save everything
AUTOML_PATH = r"C:\Users\Hzaab\Desktop\MLSD project\scratch\autogluon_auc"

# -----------------------------
# Load data
# -----------------------------
train_df = pd.read_parquet(TRAIN_PATH)
test_df = pd.read_parquet(TEST_PATH)

# -----------------------------
# Train AutoML
# -----------------------------
predictor = TabularPredictor(
    label=TARGET,
    eval_metric="roc_auc",
    path=AUTOML_PATH,
    verbosity=2,
).fit(
    train_data=train_df,
    presets="best_quality",
    time_limit=3600,   # 1 hour; change as needed
)

# -----------------------------
# Compare models
# -----------------------------
leaderboard = predictor.leaderboard(test_df, silent=True)
print("\n=== Leaderboard ===")
print(leaderboard)

# -----------------------------
# Evaluate on held-out test set
# -----------------------------
test_scores = predictor.evaluate(test_df)
print("\n=== Test metrics ===")
print(test_scores)

# -----------------------------
# Best model predictions
# -----------------------------
y_pred = predictor.predict(test_df.drop(columns=[TARGET]))
print("\nSample predictions:")
print(y_pred.head())

# Optional probabilities
y_prob = predictor.predict_proba(test_df.drop(columns=[TARGET]))
print("\nSample predicted probabilities:")
print(y_prob.head())
```

```python
import numpy as np
from sklearn.metrics import f1_score
TARGET="fake"
val_sub=pd.read_parquet(r"C:\Users\Hzaab\Desktop\MLSD project\data\preprocessed\val.parquet")
val_probs = predictor.predict_proba(val_sub.drop(columns=[TARGET]))[1]
val_y = val_sub[TARGET]


thresholds = np.linspace(0, 1, 201)

best_thresh = 0.5
best_f1 = -1

for t in thresholds:
    preds = (val_probs >= t).astype(int)
    score = f1_score(val_y, preds)
    
    if score > best_f1:
        best_f1 = score
        best_thresh = t

print("Best threshold:", best_thresh)
print("Best validation F1:", best_f1)
```
